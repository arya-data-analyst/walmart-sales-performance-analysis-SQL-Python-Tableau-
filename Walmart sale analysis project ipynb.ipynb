```python
#replace a $ sign in blank
df['unit_price']=(df['unit_price'].astype(str).str.replace('$', '',regex=False).str.strip().astype(float))

df.head()
```

```python
df.dtypes
```

```python
df.columns
```

```python
df['total']=df['unit_price']*df['quantity']
df.head()
```

```python
#total sales (revenue)
total_sales = df['total'].sum()
total_sales
```

```python
#total profit
df['profit'] = df['total'] * df['profit_margin']
total_profit = df['profit'].sum()
total_profit
```

```python
#average order value
average_order_value = df['total'].mean()
average_order_value
```

```python
#total quantity Sold
total_quantity = df['quantity'].sum()
total_quantity
```

```python
#sales_by_Branch 
sales_by_Branch = (
    df
    .groupby('Branch')['total']
    .sum()
    .reset_index()
)
print(sales_by_Branch)
```

```python
#sales_by_City
sales_by_City = (
    df
    .groupby('City')['total']
    .sum()
    .reset_index()
)
print(sales_by_City)
```

```python
#sales_by_category
sales_by_category =(
    df
    .groupby('category')['total']
    .sum()
    .reset_index()
)
print(sales_by_category)
```

```python
#profit_by_category
profit_by_category = (
        df
        .groupby('category')['profit']
        .sum() 
        .reset_index()
)
print(profit_by_category)
```

```python
# Best selling category
best_category = (
    df
    .groupby('category')['quantity']
    .sum()
    .sort_values(ascending=False)
)
best_category
```

```python
#paymentb method distribution
payment_kpi = (
    df['payment_method']
    .value_counts()
    .reset_index()
    .rename(columns={
        'index':'payment_method',
        'payment_method':
 'transactions'
    })
)

payment_kpi
```

```python
#average rating by category
avg_rating_category = (
    df
    .groupby('category',
as_index=False)['rating']
    .mean()
    .rename(columns={'rating':
'avr_rating'})
)

avg_rating_category
```

```python
#sales trend by months

df['date'] = pd.to_datetime(df['date'],errors='coerce',dayfirst=True)

df['month']= df['date'].dt.to_period('M')

monthly_sales_trend = (
    df
    .groupby('month', as_index=False)
['total']
    .sum()
)

monthly_sales_trend['month'] = monthly_sales_trend['month'].astype(str)
monthly_sales_trend

top_5_months = (
    monthly_sales_trend
     .sort_values(by='total',
  ascending=False)
    .head(5)
)

top_5_months
```

```python
#peak sales trend
df['date'] = pd.to_datetime(df['date'])

df['hour'] = df['date'].dt.hour

hourly_sales_trend = (
    df
    .groupby('hour', as_index=False)
['total']
    .sum()
)

top_5_hours = (
hourly_sales_trend.sort_values(by='total',ascending=False).head(5)
)

print(top_5_hours)
```
